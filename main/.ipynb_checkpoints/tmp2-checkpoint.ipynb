```python
#inside point
from algorithm.grapwalk_functions_v1 import count_container_single, get_type_from_list, deduplicate_union
from query.query_functions import get_all_triplets_from_o_excludes_ronly, get_all_triplets_from_ro_excludes, after_path_exclude


relation_count_container = None

for seed_item in record["seed"]:
    # print(seed_item)
    all_triples = get_all_triplets_from_o_excludes_ronly(seed_item, exclude_paths, )
    # print(len(all_triples))
    triples_count_container = count_container_single(all_triples, relation_excludes, node_excludes_funcs)
    relation_count_container = basic_union(relation_count_container, triples_count_container)
# print(overall_count_container.values())
# max_conn = max(list(relation_count_container.values()))
# print(max_conn)
common_relations_tuple = list(filter(lambda x: x[1] == len(record["seed"]), relation_count_container.items()))
common_relations = [item[0] for item in common_relations_tuple]

insidepoint_paths = []
query_error = False
for common_rel in tqdm(common_relations):

    common_relation_count_container = None
    for seed_item in record["seed"]:
        # print("Yo:", seed_item, common_rel)
        inside_points_raw = after_path_exclude(get_all_triplets_from_ro_excludes(seed_item, common_rel), exclude_paths)
        # print("Res: ", inside_points_raw)
        if inside_points_raw is None:
            query_error = True
            break
        inside_points = [item["y"]["value"] for item in inside_points_raw]

        inside_point_deduplicated_container = None

        for i_point in inside_points:
            # ro_endpoints_raw = get_all_triplets_from_s_excludes(i_point, node_excludes=exclude_paths)
            ro_endpoints_raw = get_all_triplets_from_s_excludes(i_point)
            ro_endpoints = [[item["x"]["value"], item["y"]["value"]] for item in ro_endpoints_raw]
            inside_point_type = get_type_from_list(ro_endpoints, i=0)

            insidepoint_count_container = count_container(ro_endpoints_raw, relation_excludes, node_excludes_funcs)
            inside_point_deduplicated_container = deduplicate_union(inside_point_deduplicated_container,
                                                                    insidepoint_count_container)

        common_relation_count_container = basic_union(common_relation_count_container,
                                                      inside_point_deduplicated_container)
    if query_error:
        continue
    # max_conn = max(list(common_relation_count_container.values()))
    valid_insidepoint_paths = list(filter(lambda x: x[1] == len(record["seed"]), common_relation_count_container.items()))
    insidepoint_paths += [[common_rel, "?ip"] + item[0].split("@@@") for item in valid_insidepoint_paths]
```
