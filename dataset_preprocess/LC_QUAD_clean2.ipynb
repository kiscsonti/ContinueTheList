```python
ds_lc_quad = "/home/kardosp/continuethelist/datasets/LC-QuAD/"
ds_lc_quad_train = "/home/kardosp/continuethelist/datasets/LC-QuAD/lc_quad_dump_valid_train.json"
ds_lc_quad_test = "/home/kardosp/continuethelist/datasets/LC-QuAD/lc_quad_dump_valid_test.json"
ds_qald = "/home/kardosp/continuethelist/datasets/QALD/"
ds_qald_train = "/home/kardosp/continuethelist/datasets/QALD/qald9_dump_w_answers_train.json"
ds_qald_test = "/home/kardosp/continuethelist/datasets/QALD/qald9_dump_w_answers_test.json"
```

```python
import os
import json
from copy import copy

with open(ds_lc_quad_train, "r") as f:
    lcquad_train = json.load(f)

with open(ds_lc_quad_test, "r") as f:
    lcquad_test = json.load(f)
```

```python
def fix_ds(ds):

    fixed_ds = list()

    for task in ds:

        task_copy = copy(task)
        new_res = list()
        for item in task_copy["result"]:
            if len(item.strip()) < 3:
                continue
            int_flag = False
            try:
                int(item)
                int_flag = True
            except Exception as e:
                pass

            if not int_flag:
                new_res.append(item)

        task_copy["result"] = new_res

        fixed_ds.append(task_copy)
    return fixed_ds
```

```python
fixed_lcquad_train = fix_ds(lcquad_train)
fixed_lcquad_test = fix_ds(lcquad_test)
```

```python
a_len = 0
b_len = 0

for a, b in zip(lcquad_train, fixed_lcquad_train):
    print(len(a["result"]), len(b["result"]))
    a_len += len(a["result"])
    b_len += len(b["result"])

print(a_len, b_len, "(Diff:", a_len - b_len, ")")
```

```python
a_len = 0
b_len = 0

for a, b in zip(lcquad_test, fixed_lcquad_test):
    print(len(a["result"]), len(b["result"]))
    a_len += len(a["result"])
    b_len += len(b["result"])

print(a_len, b_len, "(Diff:", a_len - b_len, ")")
```
