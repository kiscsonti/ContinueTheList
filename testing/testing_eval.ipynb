```python
import string
import random


def get_random_string(N=7):
    N = 7
    res = ''.join(random.choices(string.ascii_uppercase +
                                 string.digits, k=N))
    return res
```

```python
from sklearn.model_selection import train_test_split
from copy import copy

def get_task(ds, seed_size=5, random_seed=123):

    ds_seed = list()
    ds_eval_set = list()

    random.seed(random_seed)

    extended_ds = list()

    for task in ds:

        extended_task = copy(task)
        selection_train, selection_test = train_test_split(extended_task["result"], train_size=5)
        ds_seed.append(selection_train)
        ds_eval_set.append(selection_test)
        extended_task["task_seed"] = selection_train
        extended_task["task_gold"] = selection_test
        extended_ds.append(extended_task)
    return extended_ds
```

```python
def add_random_answer(ds, p=0.8, random_p=0.2):

    for task in ds:

        gold_choices = random.choices(task["task_gold"], k=int(p*len(task["task_gold"])))
        random_choices = [get_random_string(random.randrange(7, 15)) for _ in range(int(random_p*len(task["task_gold"])))]

        task["prediction"] = gold_choices + random_choices
    return ds
```
