```python
import os
import json
import pandas as pd
```

```python
folder = "/home/kardosp/continuethelist/evals"
```

```python
def load_json(path):
    with open(path, "r") as f:
        json_data = json.load(f)
    return json_data
```

```python
data_container = list()
for file in os.listdir(folder):
    current_file = os.path.join(folder, file)
    json_data = load_json(current_file)
    json_data["name"] = file
    data_container.append(json_data)
pdf = pd.DataFrame(data_container)
```
