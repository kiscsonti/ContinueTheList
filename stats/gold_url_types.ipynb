```python
def get_url(element):
    url = []
    mehet_flag = False
    for part in list(reversed(element.split("/"))):
        if mehet_flag:
            url.append(part)
        else:
            if "." in part:
                mehet_flag = True
                url.append(part)
    if len(url) == 0:
        return None
    return "/".join(list(reversed(url)))
```

```python
from collections import Counter
edge_container = list()

for i, record in enumerate(lc_quad_train):
    for triple in record["graph"]:
        for element in triple:
            if element.startswith("?"):
                continue

            url = get_url(element)
            if url is not None:
                edge_container.append(url)

edge_counter = Counter(edge_container)
edge_counter
```
