```python
import os
import json
import pandas as pd
import numpy as np
```

```python
folder = "/home/kardosp/continuethelist/evals2"
```

```python
os.listdir(folder)
```

```python
def load_json(path):
    with open(path, "r") as f:
        json_data = json.load(f)
    return json_data
```

```python
data_container = list()
data_container_all = list()
for file in os.listdir(folder):

    if "graphwalk" not in file:
        continue
    current_file = os.path.join(folder, file)
    json_data = load_json(current_file)
    json_data["name"] = file

    if "_all" in file:
        data_container_all.append(json_data)
    else:
        data_container.append(json_data)
pdf = pd.DataFrame(data_container)
pdf_all = pd.DataFrame(data_container_all)
```

```python
pdf["seed"] = pdf["name"].apply(lambda x: int(x.replace(".json", "").split("_")[-1][4:]))
pdf["sample"] = pdf["name"].apply(lambda x: int(x.replace(".json", "").split("_")[-2].replace("sample", "")))
pdf["rank_missing"] = pdf["rank_missing"]/250
pdf_grouped = pdf.groupby("sample").agg({"precision": [np.mean, np.std], "recall": [np.mean, np.std], "f1": [np.mean, np.std], "triple_match": [np.mean, np.std], "MR": [np.mean, np.std], "MRR": [np.mean, np.std], "hits@1": [np.mean, np.std], "hits@5": [np.mean, np.std], "hits@10": [np.mean, np.std], "rank_missing": [np.mean, np.std] })
```
