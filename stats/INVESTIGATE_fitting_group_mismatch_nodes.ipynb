```python
from evaluation.evaluate_script import graph_eval_ranking_first_nodes, get_result_group
from query.query_functions2 import get_result_from_triples

def same_graph_ranking(ds, pred_column="explanation"):
    difference_container = list()
    for i, record in enumerate(ds):
        print(i)
        if len(record[pred_column]) == 0:
            continue
        record_rank, query_size_single, best_explanation = graph_eval_ranking_first_nodes(record["graph"],
                                                                                          record[pred_column],
                                                                                          True)
        explanation_group = get_result_group(record_rank, record[pred_column])

        result_container = list()
        gold_nodeset = set(get_result_from_triples(record["graph"]))
        for explanation in explanation_group:
            result_container.append(get_result_from_triples(explanation))

        for expl_res, expl_query in zip(result_container, explanation_group):
            difference = set(expl_res) - gold_nodeset.intersection(expl_res)
            if len(difference) != 0:
                difference_container.append([record["graph"], gold_nodeset, difference, expl_res, expl_query])
    return difference_container
```
