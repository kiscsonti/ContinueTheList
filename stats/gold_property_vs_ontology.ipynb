```python
from collections import Counter
edge_container = list()

for i, record in enumerate(lc_quad_train):
    for triple in record["graph"]:
        edge_container.append(triple[1])

edge_counter = Counter(edge_container)
```

```python
edge_counter
```

```python
edge_counter2 = dict()
for k,v in edge_counter.items():
    if "ontology" in k:
        newk = k.replace("ontology", "") + "_ont"
        edge_counter2[newk] = v
    elif "property" in k:
        newk = k.replace("property", "") + "_prop"
        edge_counter2[newk] = v
    else:
        edge_counter2[k] = v
```
